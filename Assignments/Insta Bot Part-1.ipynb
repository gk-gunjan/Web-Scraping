```python
def search_and_open(profile):
    search=driver.find_element_by_xpath('//input[contains(@class,"XTCLo")]')
    search.clear()
    search.send_keys(profile)
    
    wait = WebDriverWait(driver,18)
    open_page= wait.until(EC.presence_of_element_located((By.CLASS_NAME,"-qQT3"))) 
    open_page.click()
    
def get_username(profile):
    search_and_open(profile)
    
    #clicking on followers button
    wait=WebDriverWait(driver,18)
    follower_btn=wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,"followers")))
    follower_btn.click()
    time.sleep(4)
    
    followers_list=list()
    follower_body=driver.find_element_by_xpath('//div[@class="isgrP"]')
    while len(followers_list)<500:
        #extracting name using beautifulsoup
        html_data=driver.page_source
        data=BeautifulSoup(html_data,'html.parser')
        new_data=data.find_all(class_='d7ByH')
        for i in new_data[len(followers_list):]:
            if i.a.string not in followers_list: 
                followers_list.append(i.a.string)
        #scrolling
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', follower_body)
    return followers_list   
```

```python
user=get_username('foodtalkindia')
print('The usernames of the first 500 followers of foodtalkindia are:')
for i in range(500):
    print(i+1,'---',user[i])
```

### 7 7. Checking status of story
This function can be applied on any Instagram handle.
(except private accounts not followed by you)

```python
def story_status(profile):
    search_and_open(profile)
    time.sleep(5)
    
    try:
        st=driver.find_element_by_xpath("//div[contains(@class,'h5uC0')]")
        
        h=int(driver.find_element_by_class_name("CfWVH").get_attribute('height'))
        #  Height of canvas helps to determine status of story
        if (h == 137 or h==252):
            st.click()
            print("Seeing the story right now")
        else:
            print("Story on this page is already seen by you")

    except NoSuchElementException:
        print("No story is uploaded on this page")
```
